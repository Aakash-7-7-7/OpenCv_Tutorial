## Face Detection with Haar Cascades

```python
import cv2 as cv 

img=cv.imread('image.png')
small=cv.resize(img,(500,500))
#cv.imshow('people',small)
gray=cv.cvtColor(small,cv.COLOR_BGR2GRAY)

haar_cascade=cv.CascadeClassifier('haarcascade_frontalcatface.xml')
faces_rect=haar_cascade.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=2)
print(f'no of faces found {len(faces_rect)}')

for (x,y,w,h) in faces_rect:
    cv.rectangle(small,(x,y),(x+w,y+h),(0,255,0),thickness=3)
cv.imshow('detected',small)


cv.waitKey(0)
cv.destroyAllWindows()
```
