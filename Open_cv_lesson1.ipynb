## Reading an Image

```python
import cv2 as cv 

img=cv.imread('test_image.jpg')

small=cv.resize(img,(300,200))
large=cv.resize(img,(1000,700))

cv.imshow('Small_image',small)
cv.imshow('Large',large)
cv.waitKey(0)
cv.destroyAllWindows()
```
