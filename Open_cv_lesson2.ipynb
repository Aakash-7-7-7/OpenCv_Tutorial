## Color Spaces

```python
import cv2 as cv 
import numpy as np 

img=cv.imread('test_image.jpg')
small=cv.resize(img,(900,500))
#cv.imshow('lion',small)

##--------- BGR to GRAY ----------------#
gray=cv.cvtColor(small,cv.COLOR_BGR2GRAY)
cv.imshow('gray',gray)

##----------- BGR to HSV ---------------#
hsv=cv.cvtColor(small,cv.COLOR_BGR2HSV)
cv.imshow('hsv',hsv)

##----------- BGR to LAB ---------------#
lab=cv.cvtColor(small,cv.COLOR_BGR2LAB)
cv.imshow('Lab',lab)

##----------- BGR to RGB ---------------#
bgtog=cv.cvtColor(small,cv.COLOR_BGR2RGB)
cv.imshow('BGTORG',bgtog)
cv.waitKey(0)
cv.destroyAllWindows()
```

## Color Channels

```python
import cv2 as cv
img=cv.imread('test_image.jpg')
small=cv.resize(img,(900,500))

b,g,r=cv.split(small)
cv.imshow('og',small)
cv.imshow('blue',b)
cv.imshow('green',g)
cv.imshow('red',r)

cv.waitKey(0)
cv.destroyAllWindows()
```

## Blurring Methods

```python
import cv2 as cv 

img=cv.imread('test_image.jpg')
small=cv.resize(img,(900,500))

#### Blurr
blurr=cv.blur(small,(3,3))
cv.imshow('blur',blurr)

#### Gaussian Blur
gb=cv.GaussianBlur(small,(3,3),0)
cv.imshow('GaussianBlur',gb)

#### Median Blur
mb=cv.medianBlur(small,3)
cv.imshow('Median Blur',mb)

#### Bilateral
bl=cv.bilateralFilter(small,5,15,15)
cv.imshow('BL',bl)
cv.waitKey(0)
cv.destroyAllWindows()
```
