```python
import cv2 as cv

people = ['Bradd Pitt', 'Thor', 'TonyStark']

# Load Haar cascade
haar_cascade = cv.CascadeClassifier(
    cv.data.haarcascades + 'haarcascade_frontalface_default.xml'
)

# Load trained recognizer
face_recog = cv.face.LBPHFaceRecognizer_create()
face_recog.read('face_trained.yml')

# Load test image
img = cv.imread('image copy.png')

if img is None:
    print("‚ùå Image not loaded")
    exit()

gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

# Detect face
faces_rect = haar_cascade.detectMultiScale(gray, 1.1, 3)

for (x, y, w, h) in faces_rect:
    face_roi = gray[y:y+h, x:x+w]

    label, confidence = face_recog.predict(face_roi)

    # Confidence threshold
    if confidence < 100:
        name = people[label]
    else:
        name = "Unknown"

    print(f"Prediction: {name}, Confidence: {confidence}")

    cv.rectangle(img, (x,y), (x+w,y+h), (0,255,0), 2)
    cv.putText(
        img,
        name,
        (x, y-10),
        cv.FONT_HERSHEY_COMPLEX,
        1.0,
        (0,255,0),
        2
    )

cv.imshow("Result", img)
cv.waitKey(0)
cv.destroyAllWindows()
```
